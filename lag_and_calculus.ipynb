### **General Case**

The eigenvalue problem arises when we want to find special vectors of a linear operator or a matrix $A$ that are only scaled (not rotated) when the operator acts on them. Formally,  

$$
A\mathbf{v} = \lambda \mathbf{v}
$$  

Here $\mathbf{v}$ is the eigenvector and $\lambda$ is the eigenvalue. Not all vectors behave like this, only specific directions do. Those directions define the eigenbasis of $A$, provided $A$ is diagonalizable.  

The reason eigenvalues are important is that they **simplify transformations**: once a matrix is diagonalized, it is equivalent to simple scalings on its eigenvectors.  

For Hermitian (symmetric) matrices, the eigenvalues are always real and eigenvectors form an orthonormal set. This property is essential in physics because it ensures that measured values are real and states remain orthogonal.  

### **Connection to Quantum Mechanics**

In quantum mechanics, the idea of eigenvalues and eigenvectors is fundamental. Every observable (like position, spin, momentum, or energy) is represented by a Hermitian operator $\hat{A}$.  

When we measure an observable, the only possible outcomes are its eigenvalues. The system’s wavefunction collapses into the corresponding eigenstate. Mathematically, this is expressed as  

$$
\hat{A}\ket{\psi} = a \ket{\psi}
$$  

where $a$ is the measurement outcome.  

A very central case is the **time-independent Schrödinger equation**:  

$$
\hat{H}\ket{\psi} = E \ket{\psi}
$$  

Here $\hat{H}$ is the Hamiltonian operator, $E$ are the energy eigenvalues, and $\ket{\psi}$ are the stationary states. These states evolve in time only by a phase factor, meaning the probability densities remain constant. Thus, eigenvalue problems are directly tied to the **quantization of energy levels**.  

### **Worked Example**

Let’s consider a spin-1/2 particle in a magnetic field. Its Hamiltonian is  

$$
H = a\sigma_x + b\sigma_z =
\begin{pmatrix}
b & a \\
a & -b
\end{pmatrix}
$$  

where $\sigma_x$ and $\sigma_z$ are Pauli matrices. The constants $a$ and $b$ depend on the magnetic field components.  

### **Step 1: Characteristic Equation**

To find eigenvalues, solve  

$$
\det(H - \lambda I) = 0
$$  

That is  

$$
\det
\begin{pmatrix}
b-\lambda & a \\
a & -b-\lambda
\end{pmatrix} = 0
$$  

Expanding:  

$$
(b-\lambda)(-b-\lambda) - a^2 = -(\lambda^2 - (a^2+b^2)) = 0
$$  

Therefore  

$$
\lambda = \pm \sqrt{a^2 + b^2}
$$  

These are the possible energy levels of the spin system.  

### **Step 2: Eigenvectors**

For a given eigenvalue $\lambda$, solve  

$$
(H - \lambda I)\vec{v} = 0
$$  

which becomes  

$$
\begin{pmatrix}
b-\lambda & a \\
a & -b-\lambda
\end{pmatrix}
\begin{pmatrix}
v_1 \\ v_2
\end{pmatrix} = 0
$$  

From the first row,  

$$
(b-\lambda)v_1 + a v_2 = 0
$$  

which gives  

$$
v_1 = \frac{-a}{b-\lambda}v_2
$$  

Choosing $v_2 = 1$, the unnormalized eigenvector is  

$$
\vec{v}_\lambda =
\begin{pmatrix}
-a \\
\lambda - b
\end{pmatrix}
$$  

Normalization gives the final eigenvector  

$$
\vec{v}_\lambda = \frac{1}{\sqrt{a^2+(\lambda-b)^2}}
\begin{pmatrix}
-a \\
\lambda - b
\end{pmatrix}
$$  

So the eigenvalues are $\lambda = \pm\sqrt{a^2+b^2}$ and eigenvectors are as above.

### **Example**

### **Find eigenvalues of a $3\times3$ matrix**

Take
$$
A=\begin{pmatrix}
2 & 1 & 0\\[4pt]
1 & 2 & 0\\[4pt]
0 & 0 & 3
\end{pmatrix}.
$$

**Step 1 — Characteristic polynomial**

Compute $\det(A-\lambda I)$:

$$
A-\lambda I=
\begin{pmatrix}
2-\lambda & 1 & 0\\[4pt]
1 & 2-\lambda & 0\\[4pt]
0 & 0 & 3-\lambda
\end{pmatrix}.
$$

Because the matrix is block-triangular, the determinant factors:

$$
\det(A-\lambda I)=(3-\lambda)\det\begin{pmatrix}2-\lambda & 1\\[4pt]1 & 2-\lambda\end{pmatrix}.
$$

**Step 2 — Simplify**

Compute the $2\times2$ determinant:

$$
\det\begin{pmatrix}2-\lambda & 1\\[4pt]1 & 2-\lambda\end{pmatrix}=(2-\lambda)^2-1.
$$

Expand and simplify:

$$
(2-\lambda)^2-1=\lambda^2-4\lambda+3=(\lambda-1)(\lambda-3).
$$

**Step 3 — Final result**

So the characteristic polynomial is

$$
\det(A-\lambda I)=(3-\lambda)(\lambda-1)(\lambda-3).
$$

The eigenvalues are

$$
\lambda=1,\quad \lambda=3,\quad \lambda=3.
$$

Hence, eigenvalues are **1** and **3** (with $\lambda=3$ of multiplicity 2).

```python
import numpy as np

# Define the matrix
A = np.array([[2, 1, 0],
              [1, 2, 0],
              [0, 0, 3]])

# Compute eigenvalues
eigenvalues, eigenvectors = np.linalg.eig(A)

print(eigenvalues)
```

### **Introduction to Differentiation**

Differentiation is one of the fundamental concepts in calculus. At its core, differentiation is the process of finding the derivative of a function. The derivative represents how a function changes as its input changes. In simple terms, it gives us the rate of change of a function with respect to one of its variables. Imagine you are driving a car: the position of the car changes with time, and if you want to know how fast you are going at a specific moment, you need the rate of change of position with respect to time. This is exactly what the derivative tells you. Mathematically, if $y=f(x)$, then the derivative is denoted as $\frac{dy}{dx}$ or $f'(x)$, and it describes the instantaneous rate of change of $y$ with respect to $x$.

The formal definition of a derivative comes from the concept of a limit. For a function $f(x)$, its derivative at a point $x$ is defined as:
$$
f'(x)=\lim_{h \to 0}\frac{f(x+h)-f(x)}{h}.
$$
This definition means that the derivative is the slope of the tangent line to the curve $y=f(x)$ at the point $x$. In contrast to the slope of a secant line (which is the slope between two points), the derivative gives us the slope at exactly one point by shrinking the distance between the points to zero. This concept of "instantaneous rate of change" is central not only in mathematics but also in physics, engineering, and many applied sciences.

Differentiation allows us to understand change in a precise and mathematical way. It is essential for describing motion, growth, decay, optimization problems, and much more. The idea is that whenever we want to know "how fast" something changes, or "how steep" something is at a given point, differentiation provides the tool. This is why it is often the starting point of calculus and a major stepping stone to the development of modern physics and applied mathematics.

---

### **Basic Rules and Formulas of Differentiation**

To work with differentiation efficiently, mathematicians developed a set of rules that make it easier to compute derivatives of complex functions without going back to the definition each time. Some of the most commonly used differentiation rules are:

1. **Constant Rule**: The derivative of a constant is zero.  
   $$
   \frac{d}{dx}[c] = 0
   $$

2. **Power Rule**: The derivative of $x^n$ is $nx^{n-1}$.  
   $$
   \frac{d}{dx}[x^n] = nx^{n-1}
   $$

3. **Sum Rule**: The derivative of a sum of functions is the sum of the derivatives.  
   $$
   \frac{d}{dx}[f(x)+g(x)] = f'(x)+g'(x)
   $$

4. **Product Rule**: If two functions are multiplied, their derivative is:  
   $$
   \frac{d}{dx}[f(x)g(x)] = f'(x)g(x)+f(x)g'(x)
   $$

5. **Quotient Rule**: For a ratio of two functions:  
   $$
   \frac{d}{dx}\left[\frac{f(x)}{g(x)}\right]=\frac{f'(x)g(x)-f(x)g'(x)}{[g(x)]^2}
   $$

6. **Chain Rule**: For composite functions:  
   $$
   \frac{d}{dx}[f(g(x))] = f'(g(x))\cdot g'(x)
   $$

In addition to these, we have derivatives of standard functions such as trigonometric functions, exponential functions, and logarithmic functions. For example, $\frac{d}{dx}[\sin(x)] = \cos(x)$, $\frac{d}{dx}[\cos(x)] = -\sin(x)$, $\frac{d}{dx}[e^x] = e^x$, and $\frac{d}{dx}[\ln(x)] = \frac{1}{x}$.

---

### **Differentiation in Physics**

Differentiation has a very powerful role in physics because physics is fundamentally about describing changes in nature. One of the simplest yet most fundamental examples is Newton’s second law of motion: $F=ma$. Here $F$ is the force, $m$ is the mass, and $a$ is the acceleration. But acceleration itself is the derivative of velocity with respect to time, and velocity is the derivative of position with respect to time. This means:
$$
a=\frac{dv}{dt}=\frac{d^2x}{dt^2}.
$$
Therefore, Newton’s law can also be written as:
$$
F=m\frac{d^2x}{dt^2}.
$$
This shows that force is related to the second derivative of position with respect to time. Such expressions make it very clear how differentiation translates physical concepts into mathematical language. Without differentiation, it would be impossible to describe motion with such accuracy.

Another example is in electricity and magnetism. The rate at which charge flows with respect to time is called current. Mathematically:
$$
I=\frac{dq}{dt}.
$$
Here $q$ is charge and $I$ is current. Similarly, in thermodynamics, the rate of change of internal energy with respect to temperature, under constant volume, gives the heat capacity. In wave mechanics, derivatives help describe how waves propagate through space and time. Thus, differentiation forms the backbone of nearly all physical laws.

### **Introduction to Integration**

While differentiation is concerned with rates of change, integration is concerned with accumulation and total quantities. In a sense, integration is the reverse process of differentiation. For example, if velocity is the derivative of position, then the integral of velocity with respect to time gives position. Similarly, if force is the derivative of momentum, then the integral of force gives momentum change. The notation for integration is:
$$
\int f(x)\,dx.
$$
This represents the area under the curve of $f(x)$ with respect to $x$.

Physically, integration is very intuitive. If you know the velocity of a car as a function of time, then integrating velocity over a certain time interval gives you the distance traveled. If you know the density of a rod as a function of its length, then integrating the density over the rod gives you its total mass. The key idea is that integration adds up infinitesimally small contributions to get a total. This is why it is also called the process of accumulation.

The most important rules of integration parallel those of differentiation. The integral of a constant $c$ is $cx$, the integral of $x^n$ is $\frac{x^{n+1}}{n+1}$ (for $n\neq -1$), and there are standard integrals for trigonometric, exponential, and logarithmic functions. For instance:
$$
\int e^x dx = e^x + C, \quad \int \frac{1}{x} dx = \ln|x| + C, \quad \int \sin(x)dx = -\cos(x)+C.
$$
Here $C$ is the constant of integration, which accounts for the fact that differentiation of any constant yields zero.

### **Differential Equations**

A differential equation is an equation that involves a function and its derivatives. These equations describe relationships where the rate of change of a quantity is related to the quantity itself. For example, population growth, radioactive decay, oscillations of a spring, and heat conduction are all naturally described by differential equations. There are two broad types: ordinary differential equations (ODEs), which involve derivatives with respect to one variable, and partial differential equations (PDEs), which involve derivatives with respect to multiple variables.

The simplest example is the equation:
$$
\frac{dy}{dx} = ky,
$$
where $k$ is a constant. This equation says that the rate of change of $y$ with respect to $x$ is proportional to $y$ itself. The solution to this is the exponential function:
$$
y(x)=Ce^{kx},
$$
where $C$ is a constant. This describes exponential growth when $k>0$ and exponential decay when $k<0$. Many real-life processes like population growth or radioactive decay follow this law.

Another common equation is Newton’s law of cooling:
$$
\frac{dT}{dt}=-k(T-T_{env}),
$$
where $T$ is the temperature of an object, $T_{env}$ is the surrounding environment temperature, and $k$ is a constant. This equation describes how objects cool down or heat up over time depending on their environment.

### **Solving a Simple Differential Equation**

Let us take the simple differential equation:
$$
\frac{dy}{dx}=2x.
$$
This equation says that the rate of change of $y$ with respect to $x$ is equal to $2x$. To solve it, we integrate both sides:
$$
\int \frac{dy}{dx}\,dx = \int 2x\,dx.
$$
This gives:
$$
y=x^2+C,
$$
where $C$ is a constant of integration. This is the family of solutions to the differential equation. If an initial condition is provided, such as $y(0)=1$, then $C=1$, and the specific solution becomes $y=x^2+1$.

### **Python Code to Solve the Same Differential Equation**

We can also solve this using Python. For example, let us solve $\frac{dy}{dx}=2x$ with initial condition $y(0)=1$.

```python
import sympy as sp

# Define variable and function
x = sp.symbols('x')
y = sp.Function('y')

# Define the differential equation dy/dx = 2x
diff_eq = sp.Eq(sp.Derivative(y(x), x), 2*x)

# Solve the differential equation
solution = sp.dsolve(diff_eq, y(x))

print("General Solution:", solution)

# Apply initial condition y(0)=1
ics_solution = sp.dsolve(diff_eq, y(x), ics={y(0): 1})
print("Solution with initial condition y(0)=1:", ics_solution)
```
